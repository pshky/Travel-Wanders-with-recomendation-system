## CONTENT BASED


```python
ratings_with_name = ratings.merge(location,on='Index')
```

```python
ratings_with_name
```

```python
num_rating_df = ratings_with_name.groupby('destination').count()['location-Rating'].reset_index()
num_rating_df.rename(columns={'location-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('destination').mean()['location-Rating'].reset_index()
avg_rating_df.rename(columns={'location-Rating':'avg_rating'},inplace=True)
avg_rating_df
```

```python
popular_df=num_rating_df.merge(avg_rating_df,on='destination')
popular_df
```

```python
popular_df = popular_df[popular_df['num_ratings']>=5].sort_values('avg_rating',ascending=False).head(10)
```

```python
popular_df = popular_df.merge(location,on='destination').drop_duplicates('destination')[['destination','city','Image-URL','num_ratings','avg_rating']]
```

```python
popular_df
```

```python
x = ratings_with_name.groupby('User-ID').count()['location-Rating'] >10
most_users = x[x].index
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(most_users)]
```

```python
y = filtered_rating.groupby('destination').count()['location-Rating'] >=2
famous_destination = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['destination'].isin(famous_destination)]
```

```python
pt = final_ratings.pivot_table(index='destination',columns='User-ID',values='location-Rating')
```

```python
pt
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
def recommend(location_name):
    #index fetch
    index = np.where(pt.index==location_name)[0][0]
    similar_items=sorted(list(enumerate(similarity_scores[index])),key=lambda X:x[1],reverse=True)[1:6]
    
    data=[]
    for i in similar_items:
        item = []
        temp_df =location[location['destination'] ==pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('destination')['destination'].values))
        item.extend(list(temp_df.drop_duplicates('destination')['city'].values))
        item.extend(list(temp_df.drop_duplicates('destination')['Image-URL'].values))
        
        data.append(item)
        return data               
```

```python
begnas_lake = recommend('Begnas Lake')
begnas_lake = pd.DataFrame(begnas_lake)
begnas_lake
```

```python
zoo = recommend('Zoo')
zoo = pd.DataFrame(zoo)
zoo
```

```python
result_data = pd.concat([begnas_lake, zoo])
result_data
```

```python
result_data.columns = ['Destination','City','ImageURL']
result_data
```

```python
result_data['Destination','City','ImageURL'] = result_data['Destination','City','ImageURL'].astype(float)
result_data
```

```python
result_data = result_data.reset_index(drop=True)
```
